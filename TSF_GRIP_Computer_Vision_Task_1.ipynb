**Object Detection / Optical Character Recognition (ORC)**
Character detector which extracts printed or handwritten text from an image or video.

```python
pip install -q keras-ocr
```

Import Libraries .

```python
import keras_ocr
import matplotlib.pyplot as plt
```

Initialize Pipeline.

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

Load Images.

```python
images = [
          keras_ocr.tools.read(img) for img in ['a3.jfif',
                                                'a4.jfif'
          ]
]
```

Character Extraction.

```python
predictions_group = pipeline.recognize(images)
```

```python
fig, axs = plt.subplots(nrows = len(images), figsize=(10,20))
for ax, image, predictions in zip(axs,images,predictions_group):

  keras_ocr.tools.drawAnnotations(image=image,
                                  predictions=predictions,
                                  ax=ax
                                  )
```

```python
count=1
for x in predictions_group:
  print('--------{} images--------'.format(count))
  for i in x:
    print(i[0])
  count+=1
```
